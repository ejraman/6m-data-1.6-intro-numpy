```python
import pandas as pd

# [EXERCISE] Starter Code
fruit_data = {
    "fruit": ["Apple", "Banana", "Cherry", "Date"],
    "quantity": [100, 40, 200, 20]
}

# 1) Create DataFrame
inventory = pd.DataFrame(fruit_data)

# 2) Add price column
inventory["price"] = [3.0, 4.0, 2.0, 5.0]

# 3) Calculate total_value = quantity * price
inventory["total_value"] = inventory["quantity"] * inventory["price"]

# 4) Flag low_stock = True if quantity < 50
inventory["low_stock"] = inventory["quantity"] < 50

# Display result
inventory
```

```python
import pandas as pd

# --- Setup Data (creates df) ---
df = pd.DataFrame(
    {
        "one":   [0.0, 1.0, 2.0, 3.0],
        "two":   [4.0, 5.0, 6.0, 7.0],
        "three": [8.0, 9.0, 10.0, 11.0],
        "four":  [12.0, 13.0, 14.0, 15.0],
    },
    index=["Ohio", "Colorado", "Utah", "New York"]
)

# --- [EXERCISE] Write your code here ---

# 1) Utah and New York (select those rows)
utah_newyork = df.loc[["Utah", "New York"]]

# 2) 'two' for 'Ohio'
ohio_two = df.loc["Ohio", "two"]

# 3) Select all rows where column 'three' is greater than 5
three_gt_5 = df[df["three"] > 5]

# 4) Update the value of 'four' in the 'New York' row to be 100
df.loc["New York", "four"] = 100

# --- Display results ---
print("1) Utah and New York rows:")
print(utah_newyork, "\n")

print("2) Value of 'two' for Ohio:")
print(ohio_two, "\n")

print("3) Rows where 'three' > 5:")
print(three_gt_5, "\n")

print("4) Updated DataFrame:")
print(df)
```

```python
# Setup Data
frame = pd.DataFrame({
    "b": [4, 7, -3, 2],
    "a": [0, 1, 0, 1]
}, index=['Row1', 'Row2', 'Row3', 'Row4'])
frame
```

```python
import pandas as pd

# [EXERCISE] Starter Code
scores = pd.DataFrame({
    "Student": ["Alice", "Bob", "Charlie", "David"],
    "Math": [85, 45, 92, 60],
    "Science": [70, 55, 88, 72]
})

# 1) Sort by 'Math' descending (highest first)
scores = scores.sort_values(by="Math", ascending=False).reset_index(drop=True)

# 2) Create 'Math_Rank' using rank() (highest score = rank 1)
scores["Math_Rank"] = scores["Math"].rank(ascending=False, method="min").astype(int)

# 3) Function: Pass if >= 60 else Fail
def pass_fail(score):
    return "Pass" if score >= 60 else "Fail"

# 4) Apply function to create 'Status'
scores["Status"] = scores["Math"].apply(pass_fail)

# Display result
scores
```
